```python
number2 = '5'
times = 1

while times <= 3:
    number = input('請猜一個數字1~9: ')
    if '1' <= number <= '9':
        if number < number2:
            print('太小了，請再猜一次，你還有', 3 - times, '次機會。')
            times += 1
        elif number > number2:
            print('太大了，請再猜一次，你還有', 3 - times, '次機會。')
            times += 1
        else:
            print('恭喜猜對了！')
            break
    else:
        print('請輸入1~9之間的數字')
if times > 3:
    print('挑戰失敗，你輸了！')
```
