```python
a = int(input('請輸入X座標值'))
b = int(input('請輸入y座標值'))

if a > 0:
    if b > 0:
        print('第一象限')
    elif b < 0:
        print('第四象限')
elif a < 0:
    if b > 0:
        print('第二象限')
    elif b < 0:
        print('第三象限')
else:
    print('該數值不在四個象限範圍，可能為原點、X軸或Y軸')
```

```python
for a in range(1, 1000):
    for b in range(a, 1000):
        c = 1000 - a - b
        if c > 0 and a*a + b*b == c*c:
            max = a * b * c
            print('畢式三元數=',a,b,c)
            print('最大值=',max)
            break
    else:
        continue
    break
```
