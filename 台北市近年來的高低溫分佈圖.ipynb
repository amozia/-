```python
import plotly.graph_objs as go

dates = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
最高溫 = [19.1, 19.6, 22.1, 25.7, 29.2, 32.0, 34.3, 33.8, 31.1, 27.5, 24.2, 20.7]
最低溫 = [13.9, 14.2, 15.8, 19.0, 22.3, 24.6, 26.3, 26.1, 24.8, 22.3, 19.3, 15.6]

fig = go.Figure()

fig.add_trace(go.Scatter(x=dates, y=最高溫, mode='lines+markers', name='最高溫', line=dict(color='red')))
fig.add_trace(go.Scatter(x=dates, y=最低溫, mode='lines+markers', name='最低溫', line=dict(color='blue')))
fig.update_layout(title='台北市近年來的高低溫分佈圖',
                   xaxis_title='Month',
                   yaxis_title='Temperature (°C)')

fig.show()
```

```python
!pip install matplotlib
import turtle
import random

window = turtle.Screen()
window.setup(600, 600)
john = turtle.Turtle()
john.speed('fastest')

def star():
    for _ in range(5):
        john.forward(100)
        john.right(144)

for _ in range(5):
    john.penup()
    x = random.randint(-300, 300)
    y = random.randint(-200, 200)
    john.goto(x, y)
    john.pendown()
    star()

window.mainloop()
```
