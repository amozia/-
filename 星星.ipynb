```python
!pip install ColabTurtle
import ColabTurtle.Turtle
john = ColabTurtle.Turtle

john.initializeTurtle(initial_speed=10)
from random import randint


def draw_star():
    for i in range(5):
        john.forward(100)
        john.right(144)

for j in range(5):
    john.penup()
    x = random.randint(200, 600)
    y = random.randint(200, 600)
    john.goto(x, y)
    john.pendown()
    draw_star()
```
