```python
import csv
import matplotlib.pyplot as plt
import numpy as np

# 讀取資料 字典式查詢
def read_data_from_csv(filename):
    data = {'Continent': [], '1-9Years': [], '10-19Years': [], '20-29Years': [], '30-39Years': [], '40-49Years': [], '50-59Years': [], '60Years and Over': []}
    with open(filename, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # 跳過標頭行
        for row in reader:
            data['Continent'].append(row[0])
            data['1-9Years'].append(int(row[1]))
            data['10-19Years'].append(int(row[2]))
            data['20-29Years'].append(int(row[3]))
            data['30-39Years'].append(int(row[4]))
            data['40-49Years'].append(int(row[5]))
            data['50-59Years'].append(int(row[6]))
            data['60Years and Over'].append(int(row[7]))
    return data

filename = '來臺旅客年齡.csv'
data = read_data_from_csv(filename)

# 年齡區間
age_groups = ['1-9Years', '10-19Years', '20-29Years', '30-39Years', '40-49Years', '50-59Years', '60Years and Over']

# 每個年齡區段的位置
x = np.arange(len(age_groups))

# 柱狀圖的寬度
width = 0.15

# 創建圖表
fig, ax = plt.subplots(figsize=(14, 8))

# 定義顏色
colors = ['#084f99', '#1a68ae', '#2e7ebb', '#64aad4', '#8abfdd', '#abd0e6', '#c8dcef']

# 繪製每個大洲的柱狀圖，指定顏色
for i, continent in enumerate(data['Continent']):
    ax.bar(x + i * width, [data[age_group][i] for age_group in age_groups], width, label=continent, color=colors[i % len(colors)])

# 設置標題和標籤
ax.set_xlabel('Age Range')
ax.set_ylabel('Number of People')
ax.set_title('Number of Visitors to Taiwan by Age and Continent')
ax.set_xticks(x + width * 2.5)
ax.set_xticklabels(age_groups)
ax.legend()

# 顯示圖表
plt.tight_layout()
plt.show()

```
